```python
median_math_score = data['math_score'].median()
median_math_score
```

```python
data['math_score'] = data['math_score'].fillna(median_math_score)
```

```python
median_science_score = data['science_score'].median()
median_science_score
```

```python
data['science_score'] = data['science_score'].fillna(median_science_score)
```

```python
median_english_score = data['english_score'].median()
median_english_score
```

```python
data['english_score'] = data['english_score'].fillna(median_english_score)
```

```python
median_programming_score = data['programming_score'].median()
median_programming_score
```

```python
data['programming_score'] = data['programming_score'].fillna(median_programming_score)
```

```python
median_sports_score = data['sports_score'].median()
median_sports_score
```

```python
data['sports_score'] = data['sports_score'].fillna(median_sports_score)
```

```python
mode_age = data['age'].mode()
```

```python
data['age'] = data['age'].fillna(mode_age[0])
```

```python
data['gender'] = data['gender'].fillna("Unknown")
data['city'] = data['city'].fillna("Unknown")
data['phone'] = data['phone'].fillna("Unknown")
```

```python
data.isnull().sum()
```

```python
median_attendance_pct = data['attendance_pct'].median()
median_attendance_pct
```

```python
data['attendance_pct'] = data['attendance_pct'].fillna(median_attendance_pct)
```

```python
data['age'].isnull().sum()
```

```python
data['total_score'] = data['math_score'] + data['science_score'] + data['english_score'] + data['programming_score'] + data['sports_score']
```

```python
data['total_score'].isnull().sum()
```

```python
data['average_score'] = data['total_score'] / 5
```

```python
data['average_score'].isnull().sum()
```

```python
conditions = [
    data['average_score'] >= 90,
    data['average_score'] >= 80,
    data['average_score'] >= 70,
    data['average_score'] >= 60,
    data['average_score'] >= 50,
    data['average_score'] < 50
]

grades = ['A+', 'A', 'B', 'C', 'D', 'F']

data['grade'] = np.select(conditions, grades, default='F').astype(str)
```
